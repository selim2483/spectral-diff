```python
import os 
import torch
import numpy as np

root = '/scratchm/sollivie/datasets/hytexila/unzip'
name = os.listdir(root)[1]
path = os.path.join(root, name, f'{name}.raw')

def load(path, module):
    with open(path, 'rb') as f:
        data = module.frombuffer(f.read(), dtype=module.float32)
    return data

data_np = load(path, np)
data_torch = load(path, torch)
```
